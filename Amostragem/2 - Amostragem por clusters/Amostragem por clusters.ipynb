```python
import pandas as pd
import seaborn as sb
import statistics as st
import numpy as np
import random as rd
```

```python
df = pd.read_csv(r'C:\Users\brayn\Documents\Analise de dados\Data Science\Bases da dados\Assinaturas de jornais\Journals.csv')
df.columns = ['','Titulo','Editora','sociedade','Preço','Paginas','Charpp','Citações','Ano de fundação','Subs','Assunto']
df.head()
```

```python
df.shape
```

```python
# DIVIDIR QUANTIDADE DE CONJUNTO DE DADOS COM A QUANTIDADE DE GRUPOS QUE EU QUERO OU PRECISO.
len(df) // 6

# CRIAR VARIAVEIS COMO: 
# - LISTA VAZIA (grupos = []) --> ONDE IREMOS ADICIONAR OS CLUSTERS/GRUPOS (append()).'
# id_grupo = 0 --> ONDE IREMOS LIMITAR A QUANTIDADE DE CLUSTER/GRUPOS (IGUAL A 0).
# id_contagem = 0 --> ONDE IREMOS LIMITAR A CONTAGEM (IGUAL A 0).
grupos = []
id_grupo = 0
contagem = 0

# CRIAR ESTRUTURA DE REPETIÇÃO 'for' USANDO A FUNÇÃO .iterrows() PARA VISITAR LINHA A LINHA E
# EXECUTAR AÇÕES EM CADA VALOR (LINHA POR LINHA).
for _ in df.iterrows():
    grupos.append(id_grupo)
    contagem += 1
    if contagem > 30:
        contagem = 0
        id_grupo += 1

# VERIFICAR A QUANTIDADE DE GRUPOS FEITOS PARA QUE POSSAMOS VALIDAR OS CODIGOS QUE CRIAMOS
# DENTRO DA ESTRUTURA 'for' e iterrows().
np.unique(grupos, return_counts = True)
print(grupos)

# ADICIONAR A COLUNAS GRUPOS, PARA CONTABILIZAR CADA CONJUNTO EM SEU GRUPO
df['grupos'] = grupos
df.head()

# ESCOLHENDO DE FORMA ALEATÓRIA/RANDÔMICA O GRUPO A SER ESTUDADO
rd.seed(1)
rd.randint(0,5)

# CHAMENDO O GRUPO SELECIONADO
df_group = df[df['grupos'] == 1]
df_group.head()
```
