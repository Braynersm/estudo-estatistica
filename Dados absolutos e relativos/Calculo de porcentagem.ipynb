```python
import pandas as pd
import numpy as np
import seaborn as sb
```

```python
# CALCULO DE PERCENTUAL EM UMA VARIAVEL "SEM CONTEXTO"

data = { 'Lojas':['Loja A','Loja B','Loja C','Loja D'],
        'Fritadeira':[56, 34, 45, 51],
        'Panela de Pressão':[33,42,31,50]}
df = pd.DataFrame(data)
df
```

```python
df['%_Fritadeira'] = df['Fritadeira'] / df['Fritadeira'].sum()*100
df['%_Panela de Pressão'] = df['Panela de Pressão'] / df['Panela de Pressão'].sum()*100
df.round()
```

```python
# CALCULO DE PERCENTUAL COM FILTROS COM "CONTEXTO"

df2 = pd.read_csv(r'C:\Users\brayn\Documents\Analise de dados\Estatistica e programação\0 - Bases de dados\Outros\census.csv')
df2.head()
```

```python

# 1 - Selecionar as variaveis que serão usadas na analise.
dataset1 = df2[['sex','race','income']]

# 2 - agrupar essas variaveis(.groupby() lista dentro de lista [['...']]) e depois contabilizar.
# 3 - basicamente contabilizamos usando filtros
agrupamento_por_sex_race_income = dataset1.groupby(['sex','race','income'])['sex'].count()

# 4 - criamos um index (para consultar as variaveis).
agrupamento_por_sex_race_income.index

# 5 - selecionamos os conjuntos/combinações para contabilizar(contabilizar tanto o valor: <=50k, como tambem o valor: >50k).
agrupamento_por_sex_race_income[
    ' Female',              ' Black', ' <=50K'],agrupamento_por_sex_race_income[' Female',              ' Black',  ' >50K']

# 6 - por fim calculamos para saber o percentual da variavel levando em conta o contexto(outras variaveis).
print('Mulheres negras abaixo ou igual a 50K: ', agrupamento_por_sex_race_income[
    ' Female',              ' Black', ' <=50K'] / (1465 + 90)*100)
mulheres_acima_de_50_mil = print('Mulheres negras acima de 50K: ', agrupamento_por_sex_race_income[' Female',              ' Black',  ' >50K'] / (1465+90)*100)
mulheres_acima_de_50_mil
```
